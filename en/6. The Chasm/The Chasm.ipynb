## What Hedge can do
**Count blocks and distinguish different types.**

```python
blocks = "racecar"
counts = Counter(blocks)
print(counts)
```
Output:
```text
Counter({'r': 2, 'a': 2, 'c': 2, 'e': 1})
```

You can check the count for a specific block like so:
```python
print(counts['r'])
```
Output:
```text
2
```
You can read more about the [Counter here](https://docs.python.org/3.8/library/collections.html#collections.Counter).  

The counter is a type of dictionairy. A dictionairy is a mappging between keys and values. Each key occurs exactly once in the dictionary. The value does not have to be unique. The `Counter` maps the letter to the number of occurences.
You can read more about [dicts here](https://docs.python.org/3.8/tutorial/datastructures.html#dictionaries).
